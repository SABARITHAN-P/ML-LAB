```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
X = data.data
Y = data.target
```

```python
df = pd.DataFrame(X,columns=data.feature_names)
df['target'] = Y
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='target',data=df)
plt.title("Class distribution")
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(),cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=20,random_state=42,stratify=Y)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train,Y_train)
```

```python
basic_y_predict = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix

def print_score(Y_test,y_predict):
    print(f"The Accuracy for basic DT {accuracy_score(Y_test,y_predict)}")
    print(f"The presicion score for the basic DT {precision_score(Y_test,y_predict)}")
    print(f"The recall score for the basic DT {recall_score(Y_test,y_predict)}")
    print(f"The f1 Score for the basic DT {f1_score(Y_test,y_predict)}")

print_score(Y_test,basic_y_predict)
```

```python
def conf_matrix(Y_test,y_predict,title):
    cm = confusion_matrix(Y_test,y_predict)

    sns.heatmap(cm,annot=True,fmt='d')
    plt.title(f"{title}")
    plt.xlabel("Predicted Value")
    plt.ylabel("Actuall Value")
    plt.show()

conf_matrix(Y_test,basic_y_predict,"Basic DT")
```

```python
from sklearn.metrics import roc_curve

basic_y_prob = dt.predict_proba(X_test)[:,1]
fpr,tpr,_ = roc_curve(Y_test,basic_y_prob)

plt.plot(fpr,tpr)
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False posive rate")
plt.ylabel("True positive rate")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    "criterion":['gini','entropy'],
    "max_depth" : [None,3,5,10],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4]
}


grid = GridSearchCV(
    dt,
    params,
    cv=5,
    scoring = 'f1'
)

grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
best_dt = grid.best_estimator_
best_dt.fit(X_train,Y_train)
best_dt_pred = best_dt.predict(X_test)
```

```python
print_score(Y_test,best_dt_pred)
```

```python
conf_matrix(Y_test,best_dt_pred,"For best estimator")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train,Y_train)
```

```python
basic_rf_pred = rf.predict(X_test)
```

```python
print_score(Y_test,basic_rf_pred)
```

```python
conf_matrix(Y_test,basic_rf_pred,"Basic RF")
```

```python
rf_params = {
    "n_estimators" : [50,100,200],
    "max_depth" : [None,5,10],
    "max_features" : ['sqrt','log2'],
    "bootstrap" : [True,False]
}

grid_rf = GridSearchCV(
    rf,
    rf_params,
    cv = 5,
    scoring='f1'
)

grid_rf.fit(X_train,Y_train)
```

```python
grid_rf.best_params_
```

```python
grid_rf.best_score_
```

```python
from sklearn.model_selection import cross_val_score

dt_cv_score = cross_val_score(best_dt,X,Y,cv=5)
rf_cv_score = cross_val_score(grid_rf.best_estimator_,X,Y,cv=5)

print(f"The DT Average {dt_cv_score}")
print(f"The RF Average {rf_cv_score}")
```

```python
table = pd.DataFrame({
    "Model" : ["Decision Tree","Random Forest"],
    "Accuracy" : [dt_cv_score.mean(),rf_cv_score.mean()]
})
table
```
